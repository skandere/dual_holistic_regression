```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("data/AmesHousing.csv")
df = df.drop(columns = ["Order", "PID"])
df = df.drop(columns = [df.columns[i] for (i,x) in enumerate(df.isna().sum().values) if x>1000])
df = df.drop(columns = ["Pool Area", 'Bsmt Half Bath'])
df = df.drop(columns = ['Year Built', 'Year Remod/Add', 'Yr Sold'])
df = df.dropna().reset_index(drop = True)
```
