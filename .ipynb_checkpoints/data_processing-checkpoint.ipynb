```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("data/AmesHousing.csv")
df = df.drop(columns = ["Order"])
df = df.drop(columns = [df.columns[i] for (i,x) in enumerate(df.isna().sum().values) if x>1000])
df = df.dropna().reset_index(drop = True)
```

```python
cat = [df.columns[i] for (i,x) in enumerate(df.dtypes.values) if x == 'O']
enc = OneHotEncoder(handle_unknown='ignore',sparse=False)
mat_categorical = enc.fit_transform(df[cat])
col_categorical = enc.get_feature_names(cat)
df_categorical = pd.DataFrame(mat_categorical, columns = col_categorical)
df = df.drop(columns = cat)
df = pd.concat([df, df_categorical], axis=1)
```

```python
z_scores = stats.zscore(df)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
new_df = df[filtered_entries].reset_index(drop=True)
```
